```python
import numpy
import matplotlib.pyplot as plt



def u_gen(n_samples, k = 0.3, p = 0.1):

    result = numpy.zeros(n_samples)
    
    u  = 0.0
    dv = 0.0
    for n in range(n_samples):
        if numpy.random.rand() < p:
            dv = numpy.random.rand()
            if numpy.random.rand() < 0.5:
                dv = -dv

        u = (1.0 - k)*u + k*dv
        result[n] = u 

    return result



def clamp_u_forward(u_forward, u_turn):
    max_u_forward = numpy.minimum(1 - u_turn, 1 + u_turn)
    min_u_forward = numpy.maximum(-1 - u_turn, -1 + u_turn)

    u_forward_clamp = numpy.clip(u_forward, min_u_forward, max_u_forward)
    
    return u_forward_clamp


n_samples = 500

u_forward = u_gen(n_samples)
u_turn    = 0.1*u_gen(n_samples)

u_forward = clamp_u_forward(u_forward, u_turn)

u_right   = u_forward + u_turn
u_left    = u_forward - u_turn

plt.clf()
plt.plot(u_forward, color="red")
plt.plot(u_turn, color="blue")
plt.show()

plt.clf()
plt.plot(u_right, color="red")
plt.plot(u_left, color="blue")
plt.show()
        
```
